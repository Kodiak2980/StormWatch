```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Master survey data.csv")
```

```python
df
```

```python
data_time1 = [df['Q1- Time taken'],df['Q2- Timing taken']]
```

```python
data_accuracy1 = [df[' Q1-Accuracy'],df['Q2- accuracy']]
```

```python
data_time2=[df['Q 2.1 - time taken'],df['Q2.2 - time taken']]
```

```python
data_accuracy2= [df['Q2.1- Accuracy'],df['Q2.2- Accuracy ']]
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
ax = sns.boxplot(data = data_time1)
ax = sns.swarmplot(data=data_time1, color=".25")
plt.yticks(fontsize= 15)
ax.set_ylabel("Time taken in seconds", rotation = 90, fontsize=15, labelpad=55) # y-axis label
ax.set_xlabel("Question number", rotation = 0, fontsize=15) # x-axis label
plt.xticks(np.arange(2), ('Question 1', 'Question 2'), fontsize= 15)
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
ax = sns.boxplot(data = data_time2)
ax = sns.swarmplot(data=data_time2, color=".25")
plt.yticks(fontsize= 15)
ax.set_ylabel("Time taken in seconds", rotation =90, fontsize=15, labelpad=55) # y-axis label
ax.set_xlabel("Question number", rotation = 0, fontsize=15) # x-axis label
plt.xticks(np.arange(2), ('Question 1', 'Question 2'), fontsize= 15)
```

```python
plt.figure(figsize=(10,10))
sns.set_style("darkgrid")

plt.ylim(-10, 10)
plt.yticks(fontsize= 15)
plt.xticks(np.arange(3), ('Question 1', 'Question 2'), fontsize= 15)
_=plt.plot(np.array(data_accuracy1),'.', color = 'blue', markersize=12, alpha =0.40, label = 'Type1 error')
_=plt.plot(np.array(data_accuracy2),'.', color = '#e67e00', markersize=12,alpha =0.40,label = 'Type2 error')
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
ax = sns.boxplot(data = (df['time-taken'], df['time-taken2']))
ax = sns.swarmplot(data=(df['time-taken'], df['time-taken2']), color=".25")
plt.yticks(fontsize= 15)
ax.set_ylabel("Time taken in seconds", rotation = 90, fontsize=15) # y-axis label
ax.set_xlabel("Prototype", rotation = 0, fontsize=15) # x-axis label
plt.xticks(np.arange(2), ('Type 1', 'Type 2'), fontsize= 15)
```
